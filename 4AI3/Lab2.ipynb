**3. Clean the data**

```python
# Remove the outliers
# Check how many numbers of line age>120 and check the shape
df[df['age'] > 120]
```

```python
# Keep the data with normal age and check the shape
df = df[df['age']<120]
df.info()
```

```python
# Fill NaN with mean value
# Check info to check how may NaN value
df.info()
```

```python
# Calculate the mean value and fill NaN with the mean value and the check the info
mean_chol = df['chol'].mean()
print(mean_chol)
df['chol'].fillna(mean_chol, inplace=True)
df.info()
```

```python
# Filter out the Nan Data
df_nan = df[df.isna().any(axis=1)]
print(df_nan)
```

```python
# Remove NaN and check the shape and info
df = df.dropna(axis=0)
df.info()
print(df.shape)
```

```python
# show a columns value's counts
df['cp'].value_counts()
```

```python
df.cp.value_counts()
```

**4. Split train test dataset**

```python
# Split features vs labels
# Select the labels
from sklearn.model_selection import train_test_split
y = df['target']
y.shape
```

```python
# Select the features
x = df.drop(['target'], axis=1)
x.shape
```

```python
# Splite train and test data by sklearn
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```
